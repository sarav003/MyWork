```python
%%time
%%sql

SELECT P.Name AS 'Director', M.title, M."year"
FROM Person P 
JOIN M_Director MD ON MD.PID = P.PID 
JOIN Movie M ON M.MID = MD.MID 
JOIN M_Genre MG ON MG.MID = M.MID
JOIN Genre G ON G.GID = MG.GID 
WHERE G.Name LIKE '%Comedy%'
AND M."year" % 4 = 0;
```
