As a part of this assignment you will be implementing the ```decision_function()``` of kernel SVM, here decision_function() means based on the value return by ```decision_function()``` model will classify the data point either as positive or negative

Ex 1: In logistic regression After traning the models with the optimal weights $w$ we get, we will find the value $\frac{1}{1+\exp(-(wx+b))}$, if this value comes out to be < 0.5 we will mark it as negative class, else its positive class

Ex 2: In Linear SVM After traning the models with the optimal weights $w$ we get, we will find the value of $sign(wx+b)$, if this value comes out to be -ve we will mark it as negative class, else its positive class.

Similarly in Kernel SVM After traning the models with the coefficients $\alpha_{i}$ we get, we will find the value of 
$sign(\sum_{i=1}^{n}(y_{i}\alpha_{i}K(x_{i},x_{q})) + intercept)$, here $K(x_{i},x_{q})$ is the RBF kernel. If this value comes out to be -ve we will mark $x_{q}$ as negative class, else its positive class.

RBF kernel is defined as: $K(x_{i},x_{q})$ = $exp(-\gamma ||x_{i} - x_{q}||^2)$

For better understanding check this link: https://scikit-learn.org/stable/modules/svm.html#svm-mathematical-formulation

````python
# you can write your code here
clfr=SVC(kernel="rbf",C=100,gamma=0.001)
clfr.fit(X_train,y_train)
````

````python
incpt,SV,Weight = clfr.intercept_ , clfr.support_vectors_ , clfr.dual_coef_
SV.shape
````

````python
def Function_cv(n,diff,func_vec,Weight,gamma):
    ##https://stackoverflow.com/questions/28503932/calculating-decision-function-of-svm-manually
    X_norm = np.array([np.linalg.norm(diff[n, :])])
    func_vec = func_vec + Weight[0, n] * np.exp(-gamma*(X_norm**2))
    
    return func_vec
        
````

````python

def add_one(x):
    x.append(1)

def add_zero(x):
    x.append(0)
    
````

````python
gamma=0.001
def decision_function(x_cv,gamma):
    fun_cv=[]
    decision_func=[]
   
    for q in x_cv:
        diff = SV - q
        func_vec = 0
        for n in range(np.shape(SV)[0]):
            func_vec = Function_cv(n,diff,func_vec,Weight,gamma)
        #fun_cv=decision(func_vec,incpt,decision,fun_cv)
        func_vec = func_vec + incpt
        decision_func.append(func_vec)
        if func_vec>0:                
            add_one(fun_cv)
        else:
            add_zero(fun_cv)
        
    Funcv= np.array(fun_cv)
    return Funcv,decision_func         
        
    
````

````python
N = len(f_cv)
w = np.zeros_like(f_cv[0])
b = 0
eta0 = 1
alpha = 0.001
````

````python
def log_loss(w,b,X,Y):
    sum_log=0
    for i in range(N):
        sum_log+=Y[i]*np.log10(sigmoid(w,X[i],b)) + (1-Y[i])*np. log10(1-sigmoid(w,X[i],b))
    return -1*sum_log/N
````

````python
def sigmoid(w,x,b):
    z=np.dot(x,w.T)+b
    return 1/(1+np.exp(-z))
````

````python
def new_y_value(y_cv,y_p,y_n):
    ny_cv=[]
    for val in y_cv:   
        if val==0:
            ny_cv.append(y_n)
        else:
            ny_cv.append(y_p)
    return(np.array(ny_cv))       
````

````python
def CalibratedProb(Y):
    ##https://stackoverflow.com/questions/15719519/custom-count-nonzero-function-for-numpy-versions-below-1-6
    N_Pos = (Y != 0).sum()
    N_neg = (Y != 1).sum()
    new_Y_neg = 1 / (N_neg + 2)
    new_Y_pos = (N_Pos + 1) / (N_Pos + 2)
    return new_Y_pos, new_Y_neg
````

````python
new_Y_pos, new_Y_neg = CalibratedProb(y_cv)
print(new_Y_pos)
print(new_Y_neg)
Up_y_cv=new_y_value(y_cv,new_Y_pos,new_Y_neg)
y_new = Up_y_cv
````

````python
def weight_intercept(j,eta0,alpha,f_cv,y_cv,w,b):
    w =  ((1-((eta0*alpha)/N)) * w) + (eta0 * f_cv[j]) * (y[j] -sigmoid(w,f_cv[j],b))
    b = b + (eta0)*(y[j] - sigmoid(w,f_cv[j],b))
        
    return w,b
````

````python
def sgd(f_cv,y_cv,eta0,alpha,w,b):
    Loss = []    
    for epoch in (range(0,10)):
        for k in (range(N)):
            w,b = weight_intercept(k,eta0,alpha,f_cv,y_cv,w,b)    
        lg_loss = log_loss(w,b,f_cv,y)
        Loss.append(lg_loss)
        
    return Loss,w,b
````
