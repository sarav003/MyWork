```python
rate=pd.read_csv(r"C:\Users\sarav\Downloads\Ratings.csv\Ratings.csv")
rate.head()
```

```python
book.columns
```

```python
book=book.drop(['Image-URL-S', 'Image-URL-L'], axis=1)
book.head()
```

```python
book=book.drop_duplicates(subset=["ISBN"])
book.shape
```

```python
rate.shape
```

```python
new=rate.merge(book,on='ISBN')
new.head()
```

```python
new.shape
```

```python
bol=new.groupby('User-ID').count()['Book-Rating']>150
users=bol[bol].index
```

```python
final_user=new[new['User-ID'].isin(users)]
```

```python
bo_1=final_user.groupby('Book-Title').count()['Book-Rating']>40
final=bo_1[bo_1].index
```

```python
df=final_user[final_user['Book-Title'].isin(final)]
df.head()
```

```python
temdf=df.drop_duplicates(subset=['Book-Title'])
temdf.head()
```

```python
pivot=df.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot.head()
```

```python
pivot=pivot.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(pivot)
```

```python
pivot.shape
```

```python
import string
import random
from PIL import Image
import requests
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
def recommender(name):
  ind=np.where(pivot.index==name)[0][0]
  items=sorted(list(enumerate(cosine_sim[ind])),key=lambda u:u[1],reverse=True)[1:6]
  data=[]
  pic=[]
  for ind,k in enumerate(items):
     item=[]
     temp_df = book[book['Book-Title'] == pivot.index[k[0]]]
     item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
     item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
     item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
     data.append(item)
         
  return data
```

```python
recommender('204 Rosewood Lane')
```
